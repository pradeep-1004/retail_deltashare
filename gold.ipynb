```python
%sql
from analytics_star_schema.deltashare.customers c 
inner join analytics_star_schema.deltashare.fact_sales_0 f 
on c.customer_id = f.customer_id

select c.customer_id , f.total_price , f.product_qty , c.customer_name

group by c.customer_id , f.total_price , f.product_qty , c.customer_name
order by f.total_price desc limit 5
```

```python
%sql
-- 2 
from analytics_star_schema.deltashare.customers c 
inner join analytics_star_schema.deltashare.fact_sales_0 f 
on c.customer_id = f.customer_id

select f.total_price  , c.state

group by f.total_price  , c.state
order by f.total_price desc limit 5
```

```python
%sql
--3 
from analytics_star_schema.deltashare.fact_sales_0 f 
inner join analytics_star_schema.deltashare.sales_orders s

select f.product_name , f.total_price , s.product_qty

group by f.product_name , f.total_price , s.product_qty

order by f.total_price desc limit 5
```
