```python
df = spark.read.option('header', 'true').option('inferSchema', 'true').csv('/Volumes/databricks_simulated_retail_customer_data/v01/source_files/customers.csv')

display(df.limit(10))
```

```python
from pyspark.sql.functions import * 
from pyspark.sql.types import * 
```

```python
df = df.withColumn(
    "customer_name",
    initcap(regexp_replace("customer_name", ",", ""))
)

display(df.limit(10))
```

```python
from pyspark.sql.functions import col, from_unixtime, to_date

df = (
    df
    .withColumn("postcode", col("postcode").try_cast("int"))
    .withColumn("number", col("number").try_cast("int"))
    .withColumn(
        "valid_from",
        to_date(from_unixtime(col("valid_from").try_cast("long")))
    )
    .withColumn(
        "valid_to",
        to_date(from_unixtime(col("valid_to").try_cast("long")))
    )
)

display(df.limit(10))
```

```python
df.count()
```

```python
df_dict = df.select(countDistinct('customer_id'))

display(df_dict)
```

```python
df.write.mode('overwrite').saveAsTable('analytics_star_schema.deltashare.customers')
```
